```python
import pandas as pd
import matplotlib.pyplot as plt

Sweden = pd.read_csv('/Users/anushamohan/Desktop/DemogTrans_Sweden_1749to2015_be0101tab9utveng.csv')
Sweden.columns = Sweden.iloc[0]
Sweden = Sweden.reindex(Sweden.index.drop(0))
Sweden = Sweden.dropna(how='all', axis=1)
Sweden = Sweden.reindex(Sweden.index.drop(1))
Sweden = Sweden.head(266)
Sweden['Population'] = Sweden['Population'].str.replace(',', '')
Sweden['Population'] = Sweden['Population'].str.replace('$', '')
Sweden['Population'] = Sweden['Population'].astype(int)
Sweden['Live Births'] = Sweden['Live Births'].str.replace(',', '')
Sweden['Live Births'] = Sweden['Live Births'].str.replace('$', '')
Sweden['Live Births'] = Sweden['Live Births'].astype(int)
Sweden['Deaths'] = Sweden['Deaths'].str.replace(',', '')
Sweden['Deaths'] = Sweden['Deaths'].str.replace('$', '')
Sweden['Deaths'] = Sweden['Deaths'].astype(int)
Sweden['Birth Rate']= Sweden['Live Births'] / Sweden['Population']
Sweden['Death Rate']= Sweden['Deaths'] / Sweden['Population']


Sweden
```

```python
x_values = Sweden[' Year']
birth_rate = Sweden['Birth Rate']
death_rate = Sweden['Death Rate']
plt.plot(x_values, birth_rate, label='births')
plt.plot(x_values, death_rate, label='deaths')
plt.xlabel('Year')
plt.ylabel('Rate')
plt.legend()
plt.show()
```
