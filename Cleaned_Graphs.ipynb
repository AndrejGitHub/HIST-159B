```python
Sweden = pd.read_csv('/Users/anushamohan/Desktop/sweden.csv', thousands=',')
```

```python
Sweden.head()
```

```python
x_axis = Sweden['Year']
y_axis = Sweden['Population']
plt.plot(x_axis, y_axis)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Sweden by Year')
plt.grid(True)

plt.show()
```

```python
y_new = 1000 * Sweden['LiveBirths'] / y_axis
y_new2 = 1000 * Sweden['Deaths'] / y_axis
```

```python
plt.plot(x_axis, y_new, color='blue')
plt.plot(x_axis, y_new2, color='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Sweden by Year')
plt.grid(True)

plt.show()
```

```python
Russia = Russia.dropna()
Russia.columns.tolist()
```

```python
x_axis = Russia['Year\r'].astype(int)
y_axis = Russia['\rPopulation\r'].astype(int)
y_new = 1000000 * Russia['LiveBirths'] / y_axis
y_new2 = 1000000 * Russia['Deaths\r'] / y_axis
```
